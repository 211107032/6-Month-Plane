```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO  # âœ… Corrected line
```

```python
# from PIL import Image
# import requests
# from io import BytesIO

def load_image_from_url(url):
    response = requests.get(url)  # âœ… fixed typo
    return Image.open(BytesIO(response.content))  # âœ… fixed BytesIO

elephant_url = "https://upload.wikimedia.org/wikipedia/commons/3/37/African_Bush_Elephant.jpg"
elephant = load_image_from_url(elephant_url)

plt.figure(figsize=(6, 4))
plt.imshow(elephant)  # âœ… fixed function name: should be imshow, not image
plt.axis('off') 
plt.title('Elephant')# Optional: to hide axis
plt.show()
```
